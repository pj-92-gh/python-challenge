```python
## Set up library and read file

import pandas as pd

file_name = 'election_data.csv'

df = pd.read_csv(file_name)
```

```python
## Get a visual representation of the data grouped by the candidate

g_df = df.groupby('Candidate')

g_df.describe()
```

```python
## Start a loop to get the vote counts for each unique candidate and then store that, along with their name

unique_c = df['Candidate'].unique()

store_count = []
cand_name = []

for i in range(len(unique_c)):
    z = unique_c[i]
    g_unique = df[df['Candidate']==z]
    count = g_unique.count()
    cand_name.append(z)
    store_count.append(count)
```

```python
## Now determine the percentages as well as if the candidate is a winner or loser based on who got the most votes

cand_count = []
cand_percent = []
status = []

total_count = df.count()

for i in store_count:
    cand_count.append(i[0])
    percent = (i[0]/total_count)
    cand_percent.append(percent[0])
    
max_count = max(cand_count)

for j in cand_count:
    if max_count == j:
        sts = str('Winner')
        status.append(sts)
    else:
        sts = str('Loser')
        status.append(sts)
```

```python
## Finally, create a dataframe based on the data gathered above

summary = {'Votes': cand_count, 'Name': cand_name, 'Percent': cand_percent, 'Status': status}
final_df = pd.DataFrame(summary)
final_df
```

```python
## Output to csv

final_df.to_csv('pypoll_output.csv',index=False)
```
